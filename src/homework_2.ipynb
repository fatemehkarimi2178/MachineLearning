### Importing Needed Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import Perceptron
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

### a.

**Question:** Scatter the training data by taking into the account of the class of each data.

Load the dataset:

```python
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
```

Scatter the training data:

```python
plt.scatter(train_set.x0, train_set.x1, c = train_set.target)
plt.show()
```

### b.

**Question:** Normalize the datasets according to the training data.

```python
x_train = np.array(train_set[['x0', 'x1']])
x_test = np.array(test_set[['x0', 'x1']])

y_train = np.array(train_set['target'])

scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
```

### c. 

**Question:** Fit polynomial feature extractor with degree = 5, 10, 15, 20, 25 on the datasets. Then, using LDA in each of these cases, reduce the data to a 2-dimentional space and then use the Perceptron classifier to seperate the data. Then validate all these models.

Generate a new feature matrix consisting of all polynomial combinations of the features with degree less than or equal to the 5, 10, 15, 20, 25:

```python
x_train_5 = PolynomialFeatures(degree = 5)
x_train_5 = x_train_5.fit_transform(x_train)

x_train_10 = PolynomialFeatures(degree = 10)
x_train_10 = x_train_10.fit_transform(x_train)

x_train_15 = PolynomialFeatures(degree = 15)
x_train_15 = x_train_15.fit_transform(x_train)

x_train_20 = PolynomialFeatures(degree = 20)
x_train_20 = x_train_20.fit_transform(x_train)

x_train_25 = PolynomialFeatures(degree = 25)
x_train_25 = x_train_25.fit_transform(x_train)
```

Use LDA for dimentionality reduction of each feature matrices and then use Perceptron for classification:

#### Degree = 5

```python
LDA_5 = LinearDiscriminantAnalysis(n_components = 2)
LDA_5.fit(x_train_5, y_train)

X_TRAIN_5 = LDA_5.transform(x_train_5)

CLF_5 = Perceptron()
CLF_5.fit(X_TRAIN_5, y_train)
```

Use repeated 5 fold cross-validation for validating each of the models above:

#### Degree = 5 

```python
# Degree = 5
accuracy_train_5 = np.empty((10, 5))
accuracy_test_5 = np.empty((10, 5))

for i in range(0, 10):
  X = np.arange(0, len(x_train_5))
  kf = KFold(n_splits = 5, shuffle = True)
  j = 0
  for train, test in kf.split(X):

    LDA_5 = LinearDiscriminantAnalysis(n_components = 2)
    LDA_5.fit(x_train_5[train], y_train[train])

    X_TRAIN = LDA_5.transform(x_train_5)[train]
    X_TEST = LDA_5.transform(x_train_5)[test]
    CLF_5 = Perceptron()
    CLF_5.fit(X_TRAIN, y_train[train])

    accuracy_train_5[i][j] = accuracy_score(CLF_5.predict(X_TRAIN), y_train[train])
    accuracy_test_5[i][j] = accuracy_score(CLF_5.predict(X_TEST), y_train[test])

    j = j + 1

accuracy_train_55 = np.empty(10)
accuracy_test_55 = np.empty(10)
for i in range(0, 10):
  accuracy_train_55[i] = accuracy_train_5[i].mean()
  accuracy_test_55[i] = accuracy_test_5[i].mean()
```
